```python
# The `labelEncoderDecoder` function is a utility function that can be used to encode and decode labels.
def labelEncoderDecoder(label):
    if type(label) is list:
        listEncoder = []
        for i in range(len(label)):
            if label[i] == 'Nghe máy':
                listEncoder.append(1)
            elif label[i] == 'Không thành công':
                listEncoder.append(0)
            elif label[i] == 'Lỗi':
                listEncoder.append(2)
        return listEncoder
    elif type(label) is str or type(label) is int:
        CodetoText = ''
        if str(label) == '0':
            CodetoText = 'Không thành công'
        elif str(label) == '1':
            CodetoText == 'Nghe máy'
        elif str(label) == '3':
            CodetoText == 'Lỗi'
        return CodetoText
        
```
